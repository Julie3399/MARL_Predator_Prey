```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1) Reward Shaping

```python
# Path to the CSV file
csv_file_path_DD_DD = '/Users/julie3399/Desktop/UCL MSc/Term 2/Multi-Agent/output_csv/DDPG_DDPG_3_1.csv'
csv_file_path_DD_DD_RW = '/Users/julie3399/Desktop/UCL MSc/Term 2/Multi-Agent/output_csv/DD_DD_3_1(reward_shaping).csv'

# Load the data
df_DD_DD = pd.read_csv(csv_file_path_DD_DD)
df_DD_DD_RW = pd.read_csv(csv_file_path_DD_DD_RW)

# Define the plot size
plt.figure(figsize=(10, 7))   

# Plotting the data with better visualization settings
plt.plot(range(5000),
         df_DD_DD_RW['DDPG_DDPG_3_1_RS-close - Mean Episode Reward (Last 200 episodes)'].dropna(),
         label='DDPG_DDPG_Close', color='navy', linestyle='-', linewidth=2)
plt.plot(range(5000),
         df_DD_DD_RW['DDPG_DDPG_3_1_RS-far - Mean Episode Reward (Last 200 episodes)'].dropna(),
         label='DDPG_DDPG_Far', color='crimson', linestyle='-', linewidth=2)
plt.plot(df_DD_DD['DDPG_DDPG_3_1 - Mean Episode Reward (Last 200 episodes)'].index,
         df_DD_DD['DDPG_DDPG_3_1 - Mean Episode Reward (Last 200 episodes)'],
         label='DDPG_DDPG', color='black', linestyle='-')
# Adding title and labels with enhanced font sizes
plt.title('Training Mean Episode Reward (Last 200 episodes): 3 Predators and 1 Prey', fontsize=16)
plt.xlabel('Step', fontsize=14)
plt.ylabel('Reward', fontsize=14)

# Enhancing the legend by moving it outside the plot area
plt.legend()

# Setting grid style
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.5)

# Adjusting the layout
plt.tight_layout()

# Display the plot
plt.savefig('Training Mean Episode Reward (Last 200 episodes): 3 Predators and 1 Preys (rewards).png')
plt.show()
```

# Evaluation Plots

```python
import matplotlib.pyplot as plt

# Data for the bar chart
methods = [ 'DDPG vs DDPG (Prey Close)','DDPG vs DDPG (Prey Far)','DDPG vs DDPG (No Shaping)', 'DDPG vs DDPG','MADDPG vs MADDPG', 'DDPG vs MADDPG', 
            'MADDPG vs DDPG', 'DQN vs DQN', 'Random vs Random']
collisions = [2.505, 2.765, 2.98, 2.98, 0.899, 0.8,  3.12, 0.429, 1.315]

# Creating the bar chart
plt.figure(figsize=(10, 6))
colors = ['skyblue', 'skyblue','skyblue', 'indianred', 'indianred', 'indianred', 
          'indianred', 'grey',  'grey']
plt.bar(methods, collisions, color=colors)
plt.xlabel('Methods')
plt.ylabel('Collisions per Episode')
plt.title('Collisions per Episode for Different Methods (Predator 4 Prey 2)')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5, zorder=0)
plt.savefig('Collisions per Episode for Different Methods (Predator 4 Prey 2)')
plt.show()
```
